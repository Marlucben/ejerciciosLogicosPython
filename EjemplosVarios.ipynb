```python
#Ejercicio de Par o Impar

def EsPar():
  A = int(input( 'ingrese un numero ' ))
  resultado=""
  if (A%2)==0 :
    resultado ='El numero ', A , ' es par '
  elif (A%2)==1 :
    resultado = 'El numero ', A , ' es impar '
  else : resultado = 'Ingreso invalido '
  return resultado
```

## Ejercicios con condicionales

```python
# 1 - Escribe un programa que pida a la persona usuaria que proporcione dos números y muestre el número más grande.

numero1 = int(input("Ingrese el primer numero"))
numero2 = int(input("Ingrese el segundo numero"))

if numero1>numero2 :
  print ( f"El numero {numero1} es mas grande ")
elif  numero2>numero1:
  print ( f"El numero {numero2} es mas grande ")
```

```python
#2 - Escribe un programa que solicite el porcentaje de crecimiento de producción de una empresa e informe
#si hubo un crecimiento (porcentaje positivo) o una disminución (porcentaje negativo).

numero1 = int(input("Ingrese el porcentaje de produccion"))


if numero1 > 0 :
  print ( "Se informaq que hubo porcentaje de crecimiento positivo ")
elif numero1 < 0 :
  print ( "Se informaq que hubo porcentaje de crecimiento Negativo ")
else : print (" no hubo crecimiento ")
```

```python
# 3 - Escribe un programa que determine si una letra proporcionada por la persona usuaria es una vocal o una consonante.

letra = input ("ingrese una letra")

Consonantes = 'b,c,d,f,g,h,j,k,l,m,n,p,q,r,s,t,v,w,x,y,z'
Vocales = 'a,e,i,o,u'

if letra in Consonantes :
  print (" la letra es una consonante")
elif letra in Vocales :
  print(" La letra es una vocal ")
else : print(" ingreso invalido")
```

```python
# 4 - Escribe un programa que lea valores promedio de precios de un modelo de automóvil
# durante 3 años consecutivos y muestre el valor más alto y más bajo entre esos tres años.

# Recolectamos los precios de los 3 años
precio_ano1 = float(input('Ingrese el precio promedio del automóvil en el primer año: '))
precio_ano2 = float(input('Ingrese el precio promedio del automóvil en el segundo año: '))
precio_ano3 = float(input('Ingrese el precio promedio del automóvil en el tercer año: '))
# Determinamos el valor más alto mediante comparaciones
mayor = precio_ano1
if precio_ano2 > mayor:
  mayor = precio_ano2
if precio_ano3 > mayor:
  mayor = precio_ano3
# Determinamos el valor más bajo mediante comparaciones
menor = precio_ano1
if precio_ano2 < menor:
  menor = precio_ano2
if precio_ano3 < menor:
  menor = precio_ano3
# Mostramos los resultados
print(f'El precio más alto fue de R$ {mayor}.')
print(f'El precio más bajo fue de R$ {menor}.')
```
